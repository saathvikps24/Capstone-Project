#### 4.1 - Creating Clusters

```python
#I will be using K-means clustering for the generation of clusters
#Importing Libraries
from sklearn.cluster import KMeans
```

```python
k = 3
kmeans = KMeans(n_clusters = k)
y_predicted = kmeans.fit_predict(df[['MURDER','RAPE','KIDNAPPING_AND_ABDUCTION','KIDNAPPING_AND_ABDUCTION_OF_WOMEN_AND_GIRLS','KIDNAPPING_AND_ABDUCTION_OF_OTHERS']])
y_predicted
```

```python
df['Cluster'] = y_predicted
df
```

```python
df0 = df[df["Cluster"] == 0]
df0
```

```python
df1 = df[df["Cluster"] == 1]
df1
```

```python
df2 = df[df["Cluster"] == 2]
df2
```

```python
#Renaming Clusters
cluster_names = {0:'Peaceful Area',1:'Moderate Area',2:'Sensitive Area'}
df['Cluster'] = df['Cluster'].replace(cluster_names)
```

```python
df
```

-- As you could observe from the above codes,we have successfully created clusters based on the most unsafe crimes possible (i.e Murder, Rape , Kidnapping and Abduction, Kidnapping and Abduction of women and girls, Kidnapping and abduction of others) and the most committed crimes in India(i.e Dacoity,Burglary,Theft,Robbery).

We have created three clusters, each representing the safety level of areas.

0 - Peaceful Area - Districts/Areas which have less crime level(low number in all the crimes taken into account).

1 - Moderate Area - Districts/Areas which have neither high nor low crime level, somewhere in the middle.

2 - Sensitive Area - Districts/Areas which have very high crime level (urgent action needed in these areas).

### 4.3 Analysing the clusters

Q : What is impacting more crimes in sensitive areas?

The reason for impact of more crimes in sensitive areas is due to immigration of people from rural to urban(as we can see above from the unique variables in the 'district' column) most of the districts/areas are metropolitan cities like (Bangalore, Kolkata, Ahmedabad etc.) immigration to these areas causes crime rate to increase.

(How?) 
Lack of Employment - People move to these areas for employment, while some of them do get employed but most of them won't this causes them to commit crimes like burglary, theft, murder etc.

Lack of Education - With the problem of Immigration, there is lack of education or knowledge about basic human virtues. Lack of education leads to increase in crime against women like rape,Kidnapping and Abduction etc.

Q : What needs to be done to reduce crime?

The reason why there is more crimes in sensitive area is because of lack of rules and regulations implemented by the government, although the crimes occured in sensitive areas have high literacy rate people seem to not care of the situation and that is why strict action must be taken by the government and more force should be deployed (few either from peaceful areas and moderate areas) so that crime rate can be controlled.
